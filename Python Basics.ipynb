#Python Basics


1. What is Python, and why is it popular?

Ans- Python is a high-level, versatile programming language known for its simplicity and readability. It is popular because of its easy-to-learn syntax, extensive libraries, and wide range of applications, including web development, data analysis, artificial intelligence, and automation.

2. What is an interpreter in Python?

Ans- An interpreter in Python is a program that directly executes Python code line by line, translating it into machine-readable instructions without requiring prior compilation.

3. What are pre-defined keywords in Python?

Ans- Pre-defined keywords in Python are reserved words that have specific meanings and purposes in the language. Examples include if, else, for, while, def, class, return, and import. You can't use them as variable names.

4. Can keywords be used as variable names?

Ans- No, keywords cannot be used as variable names in most programming languages because they are reserved for specific purposes in the language.

5. What is mutability in Python?

Ans- Mutability in Python refers to whether an object's value can be changed after it is created.
Mutable objects (e.g., lists, dictionaries) can be changed.
Immutable objects (e.g., tuples, strings) cannot be changed; creating a "change" makes a new object.

 6.Why are lists mutable, but tuples are immutable?

 Ans- Lists are mutable because their elements can be changed, added, or removed. Tuples are immutable to ensure data stays fixed and safe from modification.

 7.What is the difference between “==” and “is” operators in Python ?

 Ans- The == operator checks if two values are equal. The is operator checks if two objects are the same (i.e., they have the same memory location).

 8. What are logical operators in Python?

Ans- Logical operators in Python are used to combine conditional statements. The three main logical operators are:

and: Returns True if both conditions are true.
or: Returns True if at least one condition is true.
not: Reverses the result of the condition (i.e., returns True if the condition is false).

9. What is type casting in Python?

Ans- Type casting in Python refers to the conversion of one data type to another. It can be done explicitly using functions like int(), float(), str(), etc., to convert data between types such as integers, floats, and strings.

10. What is the difference between implicit and explicit type casting?

Ans- Implicit type casting (also called type conversion) is automatically done by the compiler when a smaller data type is converted to a larger data type. Explicit type casting (also called type casting or type conversion) is manually done by the programmer to convert a larger data type to a smaller one.

Example:

Implicit: int a = 5; float b = a; (Automatically converts int to float)

Explicit: float a = 5.6; int b = (int) a; (Manually converts float to int)

11. What is the purpose of conditional statements in Python?

Ans- Conditional statements in Python are used to execute specific blocks of code based on whether a condition is true or false. They help control the flow of a program, allowing it to make decisions and perform different actions depending on the conditions.

12.  How does the elif statement work?

Ans- The elif statement in Python is used to check multiple conditions. It stands for "else if" and allows you to test different expressions if the previous if or elif conditions are false. If a condition in elif is true, it executes the associated block of code. If none of the conditions are true, the else block (if present) will execute.

13. What is the difference between for and while loops?

Ans- A for loop is used when the number of iterations is known beforehand, and it's typically used for iterating over a sequence or a range.

A while loop is used when the number of iterations is unknown and continues until a specified condition is no longer true.

 14. Describe a scenario where a while loop is more suitable than a for loop?

 Ans- A while loop is more suitable when the number of iterations is unknown and depends on a condition. For example, when you need to keep asking for user input until they provide valid data, a while loop is ideal.

7. Write a function to demonstrate the behavior of mutable and immutable arguments?

```python
def modify_values(immutable_arg, mutable_arg):
    # Attempt to modify immutable argument (string)
    immutable_arg = "New Value"

    # Modify mutable argument (list)
    mutable_arg.append(4)

    print("Immutable Argument:", immutable_arg)
    print("Mutable Argument:", mutable_arg)

# Example usage
immutable = "Original"
mutable = [1, 2, 3]

modify_values(immutable, mutable)

print("Outside function - Immutable:", immutable)
print("Outside function - Mutable:", mutable)
```

14. Write a program to reverse a string using a while loop ?

```python
# Input string
string = "hello"
reversed_string = ""

# Initialize index to the last character
index = len(string) - 1

# Loop to reverse the string
while index >= 0:
    reversed_string += string[index]
    index -= 1

print("Reversed string:", reversed_string)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop ?

```python
# Get input from the user
num = int(input("Enter a number: "))

# Initialize variables
factorial = 1
i = 1

# Calculate factorial using while loop
while i <= num:
    factorial *= i
    i += 1

print(f"The factorial of {num} is {factorial}")
```
